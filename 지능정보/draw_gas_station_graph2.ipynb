# 10주차: 인천시 구별 주유소 가격 그래프로 그리기

### glob로 파일 리스트 가져오기

```python
from glob import glob
```

```python
glob('지역*.xls')
```

```python
station_files = glob('지역*.xls')
station_files
```

### 엑셀 파일 읽기

```python
import pandas as pd 

tmp_list = []

for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2)
    tmp_list.append(tmp)
    
tmp_list[0]
```

```python
stations = pd.concat(tmp_list)
stations.shape
```

```python
stations.info()
```

```python
stations.head()
```

```python
stations_sel = stations.loc[:, ['상호', '주소', '상표', '셀프여부', '휘발유']]
stations_sel
```

```python
stations_sel.columns = ['주유소', '주소', '상표', '셀프여부', '가격']
stations_sel
```

### '군/구' 정보 추출하기

```python
# '연수구'와 같은 군/구 정보만 추출하기
# '인천 계양구 벌말로 295' ->  '계양구'

gu_list = []

for address in stations['주소']:
    address_gu = address.split()[1]
    gu_list.append(address_gu)

stations_sel['군/구'] = gu_list
stations_sel
```

```python
# 군/구 목록 확인

stations_sel['군/구'].unique()
```

### 가격 정보 처리하기

```python
# 예외 처리하기

stations_sel.query("가격 == '-'")
```

```python
stations_sel = stations_sel.query("가격 != '-'")
stations_sel
```

```python
# 가격을 숫자(int)으로 변경하기

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
# SettingWithCopyWarning을 해결하기 위해 .copy() 이용

stations_sel = stations_sel.query("가격 != '-'").copy()
stations_sel
```

```python
# 가격을 숫자(int)으로 변경하기

price_list = []

for price in stations_sel['가격']:
    price_new = int(price)
    price_list.append(price_new)

stations_sel['가격'] = price_list
stations_sel
```

```python
stations_sel.info()
```

```python
stations_sel.reset_index(inplace=True)
stations_sel
```

```python
del stations_sel['index']
stations_sel
```

```python
stations_sel_name = stations_sel.set_index('상표')
stations_sel_name
```
